```python
import foxglove
import time
from foxglove.layouts import Layout, RawMessagesPanel, RawMessagesConfig

# Create a notebook buffer to collect messages
nb_buffer = foxglove.init_notebook_buffer()

for t in range(10*30):
    timestamp = t * 0.033
    foxglove.log("/hello", {"count": t}, log_time=int(timestamp*1e9))

# Create a layout with a single RawMessages Panel
layout = Layout(content=RawMessagesPanel(
    title="Test",
    config=RawMessagesConfig(
        diff_topic_path="",
        diff_enabled=False,
        diff_method="previous message",
        show_full_message_for_diff=False,
        topic_path="/hello"
    )
))

# Create a viewer and keep a reference to it
viewer = nb_buffer.show(layout_storage_key="test", layout=layout, force_layout=True)
# `viewer` needs to be the last expression in the cell for the widget to be displayed
viewer
```
